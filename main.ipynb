```python
import pandas as pd
import os
from six.moves import urllib
import mysql.connector as connection
```

```python
download_dir = "./data/"

os.makedirs(download_dir, exist_ok=True)

download_url = "https://raw.githubusercontent.com/mohitmahiyt/bank_dataset/main/bank.csv"

filename = os.path.basename(download_url)
print(filename)

download_file_path = os.path.join(download_dir,filename)
\
print(download_file_path)

urllib.request.urlretrieve(download_url, download_file_path)
```

```python
df=pd.read_csv(download_file_path, delimiter=";")
```
